 การคูณ Matrix  ($B.v$) คือ linear tranformation 

```
Bv = np.dot(B,np.array([[1],[1]]))
Bv
```

```
plt.plot([0,Bv[0,0]],[0,Bv[1,0]], '-g')  ## vector Bv
plt.plot([0,1],[0,1], '-r' )
plt.plot(0,0, '.b')
plt.plot([0,0],[-1,5], '^:k')  
plt.plot([-1.5,2.5],[0,0], '>:k') 
```

Determinant ของ Matrix ก็คือ ($Det(B$) คือ พื้นที่ของ unit vector ที่ถูกเปลี่ยนไป โดย Matrix นั้นๆ

```
plt.plot([0,Bv[0,0]],[0,Bv[1,0]], '-g')  # vector Bv
plt.plot([3,3],[7,0],  ':g') #วาดจุดไข่ปลา พื้นที่ของ Bv
plt.plot([3,0],[7,7],  ':g') ##วาดจุดไข่ปลา พื้นที่ของ Bv
plt.plot([0,1],[0,1],  '-r') ##vector v ([1,1])
plt.plot([1,1],[1,0],  ':r' )  ##วาดจุดไข่ปลาพื้นที่ของ v
plt.plot([1,0],[1,1],  ':r' )  ##วาดจุดไข่ปลาพื้นที่ของ v
plt.plot(0,0, '.b')
plt.plot([0,0],[-1,5], '^:k')  
plt.plot([-1.5,2.5],[0,0], '>:k') 
```
